Obesity Table : 

1) Top 5 regions with the highest average obesity levels in the most recent year(2022)

```python
query = """
    SELECT
    Region,
    ROUND(AVG(Mean_Estimate), 2) AS avg_obesity
    FROM obesity
    WHERE Year = 2022
    GROUP BY Region
    ORDER BY avg_obesity DESC
    LIMIT 5;
"""
pd.read_sql(query, conn)
```

2) Top 5 countries with highest obesity estimates

```python
query = """
    SELECT
    Country,
    ROUND(AVG(Mean_Estimate), 2) AS avg_obesity
    FROM obesity
    GROUP BY Country
    ORDER BY avg_obesity DESC
    LIMIT 5;
"""
pd.read_sql(query, conn)
```

3) Obesity trend in India over the years(Mean_estimate)

```python
query = """
    SELECT
    Year,
    ROUND(AVG(Mean_Estimate), 2) AS ind_avg_obesity
    FROM obesity
    WHERE Country = 'India'
    GROUP BY Year
    ORDER BY Year;
"""
pd.read_sql(query, conn)
```

4) Average obesity by gender

```python
query = """
    SELECT
    Gender,
    ROUND(AVG(Mean_Estimate), 2) AS avg_obesity
    FROM obesity
    GROUP BY Gender
     ORDER BY avg_obesity DESC;
"""
pd.read_sql(query, conn)
```

5) Country count by obesity level category and age group

```python
query = """
 select 
 age_group,
 obesity_level,
 count(DISTINCT Country) as country_count
 from obesity
 group by age_group,obesity_level
 order by age_group,obesity_level;
"""
pd.read_sql(query, conn)
```

6) Top 5 countries least reliable countries(with highest CI_Width) and Top 5 most consistent countries (smallest average CI_Width)

```python
query = """
SELECT * FROM (
    SELECT 
        'Least Reliable' AS category,
        Country,
        ROUND(AVG(CI_Width), 4) AS avg_ci
    FROM obesity
    GROUP BY Country
    ORDER BY avg_ci DESC
    LIMIT 5
)
UNION ALL
SELECT * FROM (
    SELECT 
        'Most Consistent' AS category,
        Country,
        ROUND(AVG(CI_Width), 4) AS avg_ci
    FROM obesity
    GROUP BY Country
    ORDER BY avg_ci ASC
    LIMIT 5
);
"""

pd.read_sql(query, conn)
```

7) Average obesity by age group

```python
query = """
 select age_group,ROUND(AVG(Mean_Estimate), 2) AS avg_obesity 
 from obesity
 group by age_group
 order by avg_obesity desc;
"""
pd.read_sql(query, conn)
```

8) Top 10 Countries with consistent low obesity (low average + low CI)over the years

```python
query = """
SELECT
  Country,
  ROUND(AVG(Mean_Estimate), 2) AS avg_obesity,
  ROUND(AVG(CI_Width), 4) AS avg_ci,
  ROUND(AVG(Mean_Estimate) + AVG(CI_Width), 4) AS consistency_score
FROM obesity
GROUP BY Country
ORDER BY consistency_score ASC
LIMIT 10;
"""
pd.read_sql(query, conn)
```

9) Countries where female obesity exceeds male by large margin (same year) 

```python
query = """
WITH gender_avg AS (
  SELECT
    Country,
    Year,
    AVG(CASE WHEN Gender = 'Female' THEN Mean_Estimate END) AS female_avg,
    AVG(CASE WHEN Gender = 'Male' THEN Mean_Estimate END) AS male_avg
  FROM obesity
  GROUP BY Country, Year
)
SELECT
  Country,
  Year,
  ROUND(female_avg, 2) AS female_avg,
  ROUND(male_avg, 2) AS male_avg,
  ROUND(female_avg - male_avg, 2) AS gap
FROM gender_avg
WHERE female_avg IS NOT NULL
  AND male_avg IS NOT NULL
  AND (female_avg - male_avg) >= 5
ORDER BY gap DESC;
"""
pd.read_sql(query, conn)
```

10) Global average obesity percentage per year

```python
query = """
SELECT
  Year,
  ROUND(AVG(Mean_Estimate), 2) AS global_avg_obesity
FROM obesity
GROUP BY Year
ORDER BY Year;
"""
pd.read_sql(query, conn)
```

8) Countries with increasing malnutrition (💡 Hint: Use MIN() and MAX() on Mean_Estimate per country to compare early vs. recent malnutrition levels, and filter where the difference is positive using HAVING.)

```python
query = """
SELECT
  Country,
  ROUND(MIN(Mean_Estimate), 2) AS early_min_malnutrition,
  ROUND(MAX(Mean_Estimate), 2) AS recent_max_malnutrition,
  ROUND(MAX(Mean_Estimate) - MIN(Mean_Estimate), 2) AS increase_amount
FROM malnutrition
GROUP BY Country
HAVING (MAX(Mean_Estimate) - MIN(Mean_Estimate)) > 0
ORDER BY increase_amount DESC;
"""
pd.read_sql(query, conn)
```

 Combined_Query : 

1) Obesity vs Malnutrition comparison by country (any 5 countries)

```python
query = """
SELECT
  o.Country,
  ROUND(AVG(o.Mean_Estimate), 2) AS avg_obesity,
  ROUND(AVG(m.Mean_Estimate), 2) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
  ON o.Country = m.Country
GROUP BY o.Country
ORDER BY avg_obesity DESC
LIMIT 5;
"""
pd.read_sql(query, conn)
```

2) Gender-based disparity in both obesity and malnutrition

```python
query = """
SELECT
  o.Gender,
  ROUND(AVG(o.Mean_Estimate), 2) AS avg_obesity,
  ROUND(AVG(m.Mean_Estimate), 2) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
  ON o.Country = m.Country
 AND o.Gender = m.Gender
GROUP BY o.Gender;
"""
pd.read_sql(query, conn)
```

3) Region-wise average estimates (Africa vs America)

```python
query = """
SELECT
  o.Region,
  ROUND(AVG(o.Mean_Estimate), 2) AS avg_obesity,
  ROUND(AVG(m.Mean_Estimate), 2) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
  ON o.Country = m.Country
WHERE o.Region IN ('Africa', 'Americas')
GROUP BY o.Region;
"""
pd.read_sql(query, conn)
```

4) Countries with obesity increasing and malnutrition decreasing

```python
query = """
SELECT 
    o.Country,
    ROUND(MAX(o.Mean_Estimate) - MIN(o.Mean_Estimate), 2) AS obesity_change,
    ROUND(MAX(m.Mean_Estimate) - MIN(m.Mean_Estimate), 2) AS malnutrition_change
FROM obesity o
JOIN malnutrition m 
    ON o.Country = m.Country
GROUP BY o.Country
HAVING 
    obesity_change > 0
    AND malnutrition_change < 0
ORDER BY obesity_change DESC;
"""
pd.read_sql(query, conn)
```

5) Age-wise trend analysis (Obesity vs Malnutrition)

```python
query = """
SELECT
  o.age_group,
  ROUND(AVG(o.Mean_Estimate), 2) AS avg_obesity,
  ROUND(AVG(m.Mean_Estimate), 2) AS avg_malnutrition
FROM obesity o
JOIN malnutrition m
  ON o.Country = m.Country
 AND o.age_group = m.age_group
GROUP BY o.age_group
ORDER BY o.age_group;
"""
pd.read_sql(query, conn)
```
